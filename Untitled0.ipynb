```python
import numpy as np
heart_rates=np.array([45,55,65,75,85])
average=np.mean(heart_rates)
highest=np.max(heart_rates)
lowest=np.min(heart_rates)
print("average heart rate:",average)
print("highest heart rate:",highest)
print("lowest heart rate:",lowest)
```
