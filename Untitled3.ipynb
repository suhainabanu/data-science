```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'x':[10,20,30,40,50],
    'y':[5,15,25,35,45],
    'z':[2,18,28,40,55]}
df= pd.DataFrame(data)
corr_matrix = df.corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1, square=True)
plt.tilte=('correlation heatmap')
plt.show()
```

```python
import numpy as np
data = [10,20,30,40,50,15,25,35,45]
datas = np.array(data)
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data,75)
IQR = Q1-Q3
lower_bound = Q1-1.5* IQR
upper_bound = Q3=1.5* IQR
outliers = [ x for x in datas
            if x < lower_bound or
             x> upper_bound]
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:",upper_bound)
print("outliers:", outliers)
```

```python
from scipy import stats
data_array= np.array(data)
z_scores= np.abs(stats.zscore(data_array))
outliers = data_array[z_scores>3]
print("outliers using z-score method:",outliers)
```

```python
import matplotlib.pyplot as plt
outlier_value = max(data)
outlier_index = data.index(outlier_value)
plt.scatter(range(len(data)),data, color='blue', label='data points')
plt.scatter([outlier_value],[outlier_index], color='red',label='outlier')
plt.xlabel('index')
plt.ylabel('values')
plt.title('scatter plot showing outlier')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
categories = ['apple','banana','apple','orange','banana','apple']
df = pd.DataFrame({'category':categories})
df['category'].value_counts().plot(kind='bar', color=['purple','violet','pink'])
plt.xlabel('category')
plt.ylabel('count')
plt.title('bar plot of categories data')
plt.show()
sns.countplot(x=df['category'], palette='pastel')
plt.title('count plot of categrical data')
plt.show()
```
