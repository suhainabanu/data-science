```python
import plotly.express as px
import numpy as np
df= pd.DataFrame({'x':np.linspace(0,10,100),
                  'y':np.sin(np.linspace(0,10,100))})
fig=px.line(df,x='x',y='y',title='intractive sine wave')
fig.show()
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
fig = px.scatter(tips, x='total_bill', y='tip', color='sex', size='size',
 title="total bill vs tip")
fig.show()
```

```python
import plotly.graph_objects as go
import seaborn as sns
tips = sns.load_dataset('tips')
fig = go.Figure(data=[go.Scatter3d(
    x=tips['total_bill'],
    y=tips['tip'],
    z=tips['size'],
    mode='markers',
    marker=dict(size=5, color=tips['total_bill'], colorscale='Viridis'))])
fig.update_layout(title="3d scatter plot of total bill, tip & size")
fig.show()
```
