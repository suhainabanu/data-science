```python
import pandas as pd
import numpy as np
data={"name":["sun","moon","then"],
      "BMI":["25","24","25"],
      "cholestral":[230,255,266]
      }

df=pd.DataFrame(data)
print('before imputation:')
print(df)
df['BMI'].fillna(df['cholestral'].median(),inplace=True)
print("\n after imputation:")
print(df)

```

```python
df.fillna(df.mode().iloc[0],
          inplace=True)
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
data={"name":["sun","moon","then"],
      "BMI":["25","24","25"],
      "cholestral":[230,255,266]
      }

df=pd.DataFrame(data)

scaler=StandardScaler()
df_scaled=df.copy()
df_scaled[['BMI','cholestral']]= scaler.fit_transform(df[['BMI','cholestral']])
print(df_scaled)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

data={"name":["sun","moon","then"],
      "BMI":["25","24","25"],
      "cholestral":[230,255,266]
      }
scaler = MinMaxScaler()
df=pd.DataFrame(data)
df_scaled=df.copy()
df_scaled[['BMI','cholestral']]= scaler.fit_transform(df[['BMI','cholestral']])
print(df_scaled)
```

```python
df_encoded=pd.get_dummies(df,columns=['name']),
drop_first=True
print(df_encoded)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder= LabelEncoder()
df['name'] = encoder.fit_transform(df['name'])
print(df)
```

```python
def performance_category(cholestral):
  if cholestral >=85:
    return "High"
  elif cholestral >=70:

    return "Medium"
  else:
    return"Low"
df["performance"]=df["cholestral"].apply(performance_category)
print(df)
```
